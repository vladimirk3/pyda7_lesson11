```python
#Задание 1
import pandas as pd
stats = pd.read_csv('ratings.csv')
stats.head()
def class_rate(row):
    if (row['rating'] < 0):
        return 'error'
    elif (row['rating'] <= 2):
        return 'low'
    elif (row['rating'] <= 4):
        return 'middle'
    else: 
        return 'high'
stats['class'] = stats.apply(class_rate, axis = 1)
stats
```

```python
#Задание 2
import pandas as pd
dk = pd.read_csv('keywords.csv')
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def geo_class(row, geodata):
    for region, cities in geodata.items():
        for city in cities:
            if (city in row['keyword']):
                return region
    return 'undefined'

dk['region'] = dk.apply(geo_class, args=(geo_data,), axis = 1)
#проверка
dk['region'].unique()
dk
```

```python
#Задание 3
import pandas as pd
years = list(range(1950,2020))
dr = pd.read_csv('ratings.csv')
dm = pd.read_csv('movies.csv')
data = dr.merge(dm)

def production_year(row, years):
    for year in years:
        if (str(year) in row['title']):
            return year
    return 1900

data['year'] = data.apply(production_year, args=(years,), axis = 1)
data
    
```

```python
data.groupby('year').mean()[['rating']].sort_values(by='rating', ascending=False).reset_index()
#data.groupby('year')['rating'].mean().sort_values(ascending=False).reset_index()
```
